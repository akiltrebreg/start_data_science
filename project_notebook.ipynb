```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Рассмотрим датасет, собранный на основе данных о "Среднемесячной номинальной начисленной заработной плате работников организаций по видам экономической деятельности в Российской Федерации за 2000-2023 гг." с сайта [Росстат](https://rosstat.gov.ru/labor_market_employment_salaries) и даных об уровне инфляции в стране с сайта [Инфляция в России](https://xn----ctbjnaatncev9av3a8f8b.xn--p1ai/%D1%82%D0%B0%D0%B1%D0%BB%D0%B8%D1%86%D1%8B-%D0%B8%D0%BD%D1%84%D0%BB%D1%8F%D1%86%D0%B8%D0%B8). <br>

Для анализа выбраны три направления экономической деятельности: <br>
- Обрабатывающие производства
- Образование
- Финансовая деятельность <br>

Дополнительно проведены расчеты реальной заработной платы и темпа роста прибыли с 2000-2023 гг. для каждого направления.

```python
df_salaries = pd.read_csv('salaries_data.csv')
```

```python
df_salaries.head()
```

Построим графики изменения номинальной заработной платы по годам.

```python
plt.figure(figsize=(15,8))
ax = sns.lineplot(data=df_salaries[df_salaries.economic_sector == 'Образование'], x='year', y='value')
ax.set(xlabel='Год', ylabel='Заработная плата, руб.', title='Среднемесячная номинальная начисленная заработная плата работников образовательной деятельности')
plt.show()
```

```python
plt.figure(figsize=(15,8))
ax = sns.lineplot(data=df_salaries[df_salaries.economic_sector == 'Обрабатывающие производства'], x='year', y='value', 
              color = 'green')
ax.set(xlabel='Год', ylabel='Заработная плата, руб.', title='Среднемесячная номинальная начисленная заработная плата работников обрабатывающих производств')
plt.show()
```

```python
plt.figure(figsize=(15,8))
ax = sns.lineplot(data=df_salaries[df_salaries.economic_sector == 'Финансовая деятельность'], x='year', y='value', 
              color = 'purple')
ax.set(xlabel='Год', ylabel='Заработная плата, руб.', title='Среднемесячная номинальная начисленная заработная плата работников финансовой деятельности')
plt.show()
```

*Для всех трех направлений наблюдается рост номильной зарплаты с 2000-2023 гг. <br>
В образовательной сфере наблюдается просадка в зарплате после 2010 года и после 2015 года, в финансовом - около 2008 года и в 2015 году. Для обрабатывающих производств наблюдается рост в номильной зарплате около 2008 года. <br>
Для всех трех направлений заметен более резкий рост зарплаты после 2020 года.*

Рассмотрим изменение реальных заработных плат.

```python
plt.figure(figsize=(15,8))
ax = sns.lineplot(data=df_salaries[df_salaries.economic_sector == 'Образование'], x='year', y='real_salary')
ax.set(xlabel='Год', ylabel='Заработная плата, руб.', title='Реальная заработная плата работников образовательной деятельности')
plt.show()
```

```python
plt.figure(figsize=(15,8))
ax = sns.lineplot(data=df_salaries[df_salaries.economic_sector == 'Обрабатывающие производства'], x='year', y='real_salary', 
              color = 'green')
ax.set(xlabel='Год', ylabel='Заработная плата, руб.', title='Реальная заработная плата работников обрабатывающих производств')
plt.show()
```

```python
plt.figure(figsize=(15,8))
ax = sns.lineplot(data=df_salaries[df_salaries.economic_sector == 'Финансовая деятельность'], x='year', y='real_salary', 
              color = 'purple')
ax.set(xlabel='Год', ylabel='Заработная плата, руб.', title='Реальная заработная плата работников финансовой деятельности')
plt.show()
```

*В целом для всех трех направлений наблюдается рост реальной зарплаты с 2000-2023 гг., а также просадка в зарплате около 2015 года. <br>
Для всех трех направлений заметен более резкий рост зарплаты после 2020 года.* <br>

Сравним зарплаты без учёта обесценивания денег с прошлогодними. Построим графики темпа роста реальных зарплат.

```python
plt.figure(figsize=(15,8))
ax = sns.lineplot(data=df_salaries[df_salaries.economic_sector == 'Образование'], x='year', y='salary_rate')
ax.set(xlabel='Год', ylabel='Темп роста, %', title='Темп роста реальной заработной платы работников образовательной деятельности')
plt.show()
```

```python
plt.figure(figsize=(15,8))
ax = sns.lineplot(data=df_salaries[df_salaries.economic_sector == 'Обрабатывающие производства'], x='year', y='salary_rate', 
              color = 'green')
ax.set(xlabel='Год', ylabel='Темп роста, %', title='Темп роста реальной заработной платы работников обрабатывающих производств')
plt.show()
```

```python
plt.figure(figsize=(15,8))
ax = sns.lineplot(data=df_salaries[df_salaries.economic_sector == 'Финансовая деятельность'], x='year', y='salary_rate', 
              color = 'purple')
ax.set(xlabel='Год', ylabel='Темп роста, %', title='Темп роста реальной заработной платы работников финансовой деятельности')
plt.show()
```

Посмотрим на минимальный и максимальный рост реальных зарплат с 2004 по 2023 год.

```python
df_salaries[df_salaries.year >= 2004].groupby('economic_sector').agg(
    min_rate=('salary_rate', np.min),
    max_rate=('salary_rate', np.max)).reset_index()
```

*Наблюдаемые на графиках снижения темпов роста реальной заработной платы относительно прошлогодней в трех экономических секторах обусловлены следующими причинами:* <br>
- *Финансово-экономический кризис в 2008—2010 гг.*
- *Валютный кризис в 2014—2015 гг.*
- *Экономический кризис 2022 года*
